```python
# Function to change time format
import pandas as pd
from datetime import datetime, time

def convert_to_normal_time(time_24):
    if isinstance(time_24, str):
        # Split the time into hours and minutes
        hours, minutes = map(int, time_24.split(':'))
    elif isinstance(time_24, pd.Timestamp):
        # Extract hours and minutes from a pandas Timestamp
        hours = time_24.hour
        minutes = time_24.minute
    elif isinstance(time_24, time):
        # Extract hours and minutes from a datetime.time object
        hours = time_24.hour
        minutes = time_24.minute
    else:
        return None  # Handle cases where the input is not valid

    # Determine AM/PM and convert hours
    period = 'AM' if hours < 12 else 'PM'
    hours = hours % 12
    hours = 12 if hours == 0 else hours  # Convert 0 hours to 12
    
    return f"{hours:02}:{minutes:02} {period}"
```

```python
# Attempt to change time to AM/PM period
Final_df['Departing_Time']=Final_df['Departing_Time'].apply(convert_to_normal_time)
Final_df['Reaching_Time']=Final_df['Reaching_Time'].apply(convert_to_normal_time)
```
